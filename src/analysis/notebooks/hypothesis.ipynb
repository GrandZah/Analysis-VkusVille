```python
import pandas as pd
```

```python
df = pd.read_csv("../data/data_clean.tsv", sep="\t")
df
```

```python
df["w_p_r"] = df["weight_g"] / df["price_rub"]
```

```python
df.loc[df["w_p_r"].nlargest(10).index]
```
