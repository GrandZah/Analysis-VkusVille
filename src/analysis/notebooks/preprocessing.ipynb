```python
df = df.drop(columns=["url", "name", "category_path", "ingredients", "tags", "image_path", "brand"])
```

```python
min_count = 10

for col in ["country", "category_main"]:
    vc = df[col].value_counts()
    frequent = vc[vc >= min_count].index

    dummies = pd.get_dummies(df[col])
    dummies = dummies.reindex(columns=frequent, fill_value=0)

    other = df[col].isin(frequent).astype(int)
    other = 1 - other
    other = other.rename(f"{col}_other")

    df = df.join(dummies.astype("uint8")).join(other.astype("uint8"))

df = df.drop(columns=["country", "category_main"])
```

```python
import re, json
from typing import List, Tuple, Dict, Any

LEGAL = ("ООО","ИП","АО","ПАО","ОАО","ЗАО","НАО","МПК","ТПК","НПО","НПФ","ПК","ТД","ТС","МП")
COLON = re.compile(r':')
LEGAL_RE = re.compile(r'\b(?:' + '|'.join(map(re.escape, LEGAL)) + r')\b', re.I)
ENDS_OOO_RE = re.compile(r'(?:["«»„“”]\s*)*\bООО\s*$', re.I)
TRIM = ' \t\r\n,.;:«»"„“”'

def _clean(x: str) -> str:
    return ' '.join(x.split()).strip(TRIM)

def _left_boundary_words_until_digit(s: str, from_idx: int) -> int:
    tokens = list(re.finditer(r'\S+', s[:from_idx]))
    for t in reversed(tokens):
        w = t.group()
        if any(ch.isdigit() for ch in w) or '.' in w or ';' in w:
            return t.end()
    return 0

def extract_names_with_logging(line: str) -> Tuple[List[str], List[Dict[str, Any]]]:
    if not isinstance(line, str) or not line:
        return (["Unknowing_manufacturer"],
                [{"case": "fallback_unknown", "colon_pos": None, "start": None, "raw": "", "name": "Unknowing_manufacturer"}])

    out, logs = [], []
    for m in COLON.finditer(line):
        ci = m.start()
        seg = line[:ci]
        last_legal = None
        for lm in LEGAL_RE.finditer(seg):
            last_legal = lm

        if last_legal:
            tail = _clean(seg[last_legal.end():])
            if tail:
                start = last_legal.start()
                case = "case_1_legal_tail_anchor_legal"
            elif ENDS_OOO_RE.search(seg[last_legal.start():]):
                start = _left_boundary_words_until_digit(line, last_legal.start())
                case = "case_2_ends_with_OOO_digit_stop"
            else:
                start = 0
                case = "case_1_legal_no_tail"
        else:
            start = _left_boundary_words_until_digit(line, ci)
            case = "case_3_no_legal_digit_stop"

        raw = line[start:ci]
        name = _clean(raw)
        logs.append({"case": case, "colon_pos": ci, "start": start, "raw": raw, "name": name})
        if name:
            out.append(name)

    seen, names, kept_logs = set(), [], []
    for n, lg in zip(out, logs):
        k = n.lower()
        if k not in seen:
            seen.add(k); names.append(n); kept_logs.append(lg)

    if not names:
        names = ["Unknowing_manufacturer"]
        kept_logs.append({"case": "fallback_unknown", "colon_pos": None, "start": None, "raw": "", "name": "Unknowing_manufacturer"})

    return names, kept_logs
```

```python
df['names_list'] = df['manufacturer'].apply(lambda s: extract_names_with_logging(s)[0])
df
```

```python
# Порог частоты: всё, что встречается реже, уедет в "other"
min_count = 10  # поставьте своё число

# 1. Разворачиваем списки
exploded = df["names_list"].explode()

# 2. Считаем частоты по всем именам
value_counts = exploded.value_counts()

# 3. Берём только те имена, что прошли порог
frequent_names = value_counts[value_counts >= min_count].index
```

```python

# 4. Делаем one-hot только по частым
multi_hot_frequent = (
    pd.get_dummies(exploded)
      .reindex(columns=frequent_names)          # оставляем только частые
      .groupby(level=0).max()
      .reindex(df.index, fill_value=0)
      .astype("uint8")
)

# 5. Делаем столбец "other": 1, если у строки есть ХОТЬ ОДНО редкое имя
other_series = (
    exploded.where(~exploded.isin(frequent_names))   # оставили только редкие, частые -> NaN
            .groupby(level=0)
            .apply(lambda s: int(s.notna().any()))   # если в группе было хоть одно редкое -> 1
            .reindex(df.index, fill_value=0)
            .astype("uint8")
            .rename("Other Manifactures")
)

# 6. Собираем всё обратно
df = df.join(multi_hot_frequent).join(other_series)
```

```python
df = df.drop(columns=["names_list", "manufacturer"])
```

```python
def robust_scale(df, cols):
    q1 = df[cols].quantile(0.25)
    q3 = df[cols].quantile(0.75)
    median = df[cols].median()
    df[cols] = (df[cols] - median) / (q3 - q1)
    return df

def minmax_scale(df, cols):
    min_val = df[cols].min()
    max_val = df[cols].max()
    df[cols] = (df[cols] - min_val) / (max_val - min_val)
    return df
```
